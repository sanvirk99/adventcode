```python
file = open('./input/day6.txt','r')


grid=[]
with file as f:
    for line in f:
        line=line.strip()
        grid.append([])
        for element in line:
            grid[-1].append(element)

print(grid)
```

```python
direction={
    'up' : [-1,0],
    'down':[1,0],
    'right':[0,1],
    'left':[0,-1]
}

original=set()
def dfs(i,j,facing):
    while True:
        if not (0 <= i < len(grid) and 0 <= j < len(grid[0])):
            return
        if grid[i][j] == '#':
            #change direction
            if facing == 'up':
                i=i+1
                facing='right'
                continue
                # dfs(i+1,j,'right')
            if facing == 'down':
                i=i-1
                facing='left'
                continue
                # dfs(i-1,j,'left')
            if facing == 'right':
                j=j-1
                facing='down'
                continue
                # dfs(i,j-1,'down')
            if facing == 'left':
                j=j+1
                facing='up'
                continue
                # dfs(i,j+1,'up')
        else:
            original.add((i,j))
            grid[i][j] = 'X'
            ix,iy = direction[facing]
            i=i+ix
            j=j+iy
            # dfs(i+ix,j+iy,facing)


def getStart():
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == '^':
                return [i,j]
            
    return [-1,-1]

sx ,sy = getStart()        
dfs(sx,sy,'up')

res=0
for row in grid:
    for x in row:
        if x == 'X':
            res+=1

print(res)
print(len(original), 'size of set')
```
